```python
def find_nearest(hospital_df, location_df, lat_col, lon_col, id_col, month_col, year_col):
    nearest = {}
    
    for _, row in hospital_df.iterrows():
        # Filter location_df by matching Month and Year
        filtered_location_df = location_df[
            (location_df[month_col] == row['Month']) & 
            (location_df[year_col] == row['Year'])
        ]
        
        # If no locations match the Month and Year, assign NaN
        if filtered_location_df.empty:
            nearest[row['ID']] = np.nan
            continue
        
        # Create a cKDTree for the filtered locations
        tree = cKDTree(filtered_location_df[[lat_col, lon_col]].values)
        
        # Find the nearest neighbor in the filtered dataset
        _, idx = tree.query([row['Transformed_Latitude'], row['Transformed_Longitude']])
        nearest_location = filtered_location_df.iloc[idx]
        
        # Add the nearest location's id_col to the result
        nearest[row['ID']] = nearest_location[id_col]

    return nearest
```

```python
train_nearest = find_nearest(train, support_data, "support_Transformed_Latitude", "support_Transformed_Longitude", "support_Month_Year_lat_lon", "support_Month", "support_Year")

train_nearest_df = pd.DataFrame(list(train_nearest.items()), columns=['ID', 'support_Month_Year_lat_lon'])

merged_train_data = train.merge(train_nearest_df, on="ID", how="left").merge(support_data, on="support_Month_Year_lat_lon", how="left")
```

```python
test_nearest = find_nearest(test, support_data, "support_Transformed_Latitude", "support_Transformed_Longitude", "support_Month_Year_lat_lon", "support_Month", "support_Year")

test_nearest_df = pd.DataFrame(list(test_nearest.items()), columns=['ID', 'support_Month_Year_lat_lon'])

merged_test_data = test.merge(test_nearest_df, on="ID", how="left").merge(support_data, on="support_Month_Year_lat_lon", how="left")
```
