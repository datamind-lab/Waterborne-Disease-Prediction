```python
train_data = train_data[train_data['Disease'] != 'Cholera']
test_data = test_data[test_data['Disease'] != 'Cholera']
```

```python
train_data = train_data.drop(columns=['ID', 'Location', 'Category_Health_Facility_UUID', 
                                      'support_Month_Year_lat_lon', 'support_Year','support_Month',
                                      'support_Transformed_Latitude', 'support_Transformed_Longitude', 
                                      'support_Month_Year', 'support_lat_lon'])

test_data = test_data.drop(columns=['ID', 'Location', 'Category_Health_Facility_UUID', 
                                      'support_Month_Year_lat_lon', 'support_Year','support_Month',
                                      'support_Transformed_Latitude', 'support_Transformed_Longitude', 
                                      'support_Month_Year', 'support_lat_lon'])
```

```python
train_data['Outbreak_Status'] = train_data.apply(
    lambda row: row['Disease'] if row['Total'] > 0 else 'No outbreaks', axis=1
)
```

```python
train_data['Outbreak_Status'].value_counts()
```

```python
label_encoder = LabelEncoder()

# Encode the 'Outbreak_Status' column
train_data['Outbreak_Status_Label'] = label_encoder.fit_transform(train_data['Outbreak_Status'])
test_data['Disease_Label'] = label_encoder.transform(test_data['Disease'])
# Show the encoded labels and their corresponding classes
print(label_encoder.classes_)
classes=label_encoder.classes_
print(label_encoder.transform(classes))
```

```python
train_data.columns
```

```python
columns_of_interest = train_data.select_dtypes(include=['number'])
correlation_matrix = columns_of_interest.corr()
threshold = 0.9
correlated_pairs = []

# Iterate over the upper triangle of the correlation matrix (to avoid duplicates)
for col in correlation_matrix.columns:
    for row in correlation_matrix.index:
        if (correlation_matrix.loc[row, col] > threshold) and (row != col):
            correlated_pairs.append((row, col, correlation_matrix.loc[row, col]))

# Create a DataFrame of the correlated pairs
correlated_pairs_df = pd.DataFrame(correlated_pairs, columns=['Feature_1', 'Feature_2', 'Correlation'])
correlated_pairs_df
```

## *Classification Model Building*

```python
x=train_data.drop(['Total','Disease','Outbreak_Status','Outbreak_Status_Label'],axis=1)
y=train_data['Outbreak_Status_Label']

x_test=test_data.drop(['Disease','Disease_Label'],axis=1)
y_test=test_data['Disease_Label']

x_train,x_val,y_train,y_val=train_test_split(x,y,random_state=1)
```

```python
x_train.describe()
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler on training data and transform both training and testing data
x_train_scaled = pd.DataFrame(scaler.fit_transform(x_train),columns=x_train.columns)
x_val_scaled = pd.DataFrame(scaler.transform(x_val),columns=x_val.columns)
x_test_scaled = pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
```

```python
x_train_scaled.describe()
```

*Decision Tree Model*

```python
dtree_model=DecisionTreeClassifier()
dtree_model.fit(x_train,y_train)
```

```python
y_pred = dtree_model.predict(x_test)

print(f"Accuracy on Training data: {round(dtree_model.score(x_train, y_train) * 100, 2)}%")
print(f"Accuracy on Validation data: {round(dtree_model.score(x_val, y_val) * 100, 2)}%")
print(f"Accuracy on Test data: {round(dtree_model.score(x_test, y_test) * 100, 2)}%")
```

```python
y_pred=dtree_model.predict(x_val)
dtree_report=classification_report(y_val,y_pred)
```

```python
print(dtree_report)
```

*Multicollinearity*

```python
importances = dtree_model.feature_importances_

# Show the feature importances along with the feature names
feature_importance = pd.DataFrame({
    'Feature': x_train.columns,
    'Importance': importances
})

# Sort the features by importance
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

feature_importance
```

*RandomForest Model*

```python
rf_model=RandomForestClassifier()
rf_model.fit(x_train,y_train)
```

```python
y_pred = rf_model.predict(x_test)

print(f"Accuracy on Training data: {round(rf_model.score(x_train, y_train) * 100, 2)}%")
print(f"Accuracy on Validation data: {round(rf_model.score(x_val, y_val) * 100, 2)}%")
print(f"Accuracy on Test data: {round(rf_model.score(x_test, y_test) * 100, 2)}%")
```

```python
y_pred=rf_model.predict(x_val)
rf_report=classification_report(y_val,y_pred)
```

```python
print(rf_report)
```

```python
train_reg_data=train_data[train_data['Outbreak_Status_Label']!=4]
```

```python
train_reg_data
```

```python
x_reg=train_reg_data.drop(['Total','Disease','Outbreak_Status'],axis=1)
y_reg=train_reg_data['Total']
```

```python
x_reg_train,x_reg_val,y_reg_train,y_reg_val=train_test_split(x_reg,y_reg,random_state=3)
```

```python
x_reg_train_scaled.describe()
```

```python
dtree_reg_model=DecisionTreeRegressor()
dtree_reg_model.fit(x_reg_train,y_reg_train)
```

```python
y_reg_pred=dtree_reg_model.predict(x_reg_val)

print(f"Accuracy on Training data: {round(dtree_reg_model.score(x_reg_train, y_reg_train) * 100, 2)}%")
print(f"Accuracy on Validation data: {round(dtree_reg_model.score(x_reg_val, y_reg_val) * 100, 2)}%")
print(f"Mean Absolute Error: {mean_absolute_error(y_reg_val,y_reg_pred)}")
```
