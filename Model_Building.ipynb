```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.utils import resample
```

```python
train_data=pd.read_csv("Merged_Train_data.csv")
test_data=pd.read_csv("Merged_Test_Data.csv")
```

```python
train_data = train_data[train_data['Disease'] != 'Cholera']
test_data = test_data[test_data['Disease'] != 'Cholera']
```

```python
train_data.shape
```

```python
train_data = train_data.drop(columns=['ID', 'Location', 'Category_Health_Facility_UUID', 
                                      'support_Month_Year_lat_lon', 'support_Year','support_Month',
                                      'support_Transformed_Latitude', 'support_Transformed_Longitude', 
                                      'support_Month_Year', 'support_lat_lon'])

test_data = test_data.drop(columns=['ID', 'Location', 'Category_Health_Facility_UUID', 
                                      'support_Month_Year_lat_lon', 'support_Year','support_Month',
                                      'support_Transformed_Latitude', 'support_Transformed_Longitude', 
                                      'support_Month_Year', 'support_lat_lon'])
```

```python
train_data['Total'].value_counts()
```

```python
train_data['Outbreak_Status'] = train_data.apply(
    lambda row: row['Disease'] if row['Total'] > 0 else 'No outbreaks', axis=1
)
```

```python
train_data['Outbreak_Status'].value_counts()
```

```python
label_encoder = LabelEncoder()

# Encode the 'Outbreak_Status' column
train_data['Outbreak_Status_Label'] = label_encoder.fit_transform(train_data['Outbreak_Status'])
test_data['Disease_Label'] = label_encoder.transform(test_data['Disease'])
# Show the encoded labels and their corresponding classes
print(label_encoder.classes_)
classes=label_encoder.classes_
print(label_encoder.transform(classes))
```

```python
train_data.shape
```

```python
train_data.columns
```

## *Classification Model Building*

```python
x=train_data.drop(['Total','Disease','Outbreak_Status','Outbreak_Status_Label'],axis=1)
y=train_data['Outbreak_Status_Label']

x_test=test_data.drop(['Disease','Disease_Label'],axis=1)
y_test=test_data['Disease_Label']

x_train,x_val,y_train,y_val=train_test_split(x,y,random_state=1)
```

```python
dtree_model=DecisionTreeClassifier()
dtree_model.fit(x_train,y_train)
```

```python
y_pred = dtree_model.predict(x_test)

print(f"Accuracy on Training data: {round(dtree_model.score(x_train, y_train) * 100, 2)}%")
print(f"Accuracy on Validation data: {round(dtree_model.score(x_val, y_val) * 100, 2)}%")
print(f"Accuracy on Test data: {round(dtree_model.score(x_test, y_test) * 100, 2)}%")
```
